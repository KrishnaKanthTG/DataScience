# 1. Read the dataset to the python environment.

```python
import pandas as pd
car=pd.read_csv('cars_data.csv')
car
```

As it is fetched, there exist 10000 rows and 20 columns.
# 2. Check for the null values present in the dataset.

```python
car.isnull().head()
```


# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
import seaborn as sns
import matplotlib.pyplot as plt
data=car
p=sns.set_theme(style="darkgrid")
sns.countplot(x="Buyer Gender",hue='Buyer Gender',data=data)
plt.legend(loc='upper center')
plt.title('Male Vs Female Buyers Participated in the Sales',fontsize=15)
plt.xlabel('Buyer Gender',fontsize=10)
plt.ylabel('Count',fontsize=10)
plt.xticks()
plt.yticks()
```

 Inference : The number of male buyers is less than female ones
 
# 4. Find the top 5 cars based on their sales price.

```python
car1=car[["Make","Model","Sale Price"]]
car1.nlargest(5,"Sale Price")
```

Based on the sales price, these above five are the top 5 cars.
# 5. Find the least 5 cars based on their Resell price.

```python
car2=car[["Make","Model","Sale Price","Resell Price"]]
car2.nsmallest(5,"Resell Price")
```
